```python
X=df[['Married','Education','ApplicantIncome','LoanAmount','Credit_History']]
```

```python
X
```

```python
Y = df[['Loan_Status']]
```

```python
Y
```

```python
from sklearn import preprocessing
lb = preprocessing.LabelEncoder()
import warnings
warnings.filterwarnings('ignore')
```

```python
Y['Loan_Status']=lb.fit_transform(Y['Loan_Status'])
```

```python
Y
```

```python
X['Education']=lb.fit_transform(X['Education'])
```

```python
X['Married']=lb.fit_transform(X['Married'])
```

```python
X
```

```python
X['Credit_History'][0]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
X
```

```python
import joblib
joblib.dump(scaler,'ScalerLA.pkl')
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3)
```

```python
 from sklearn.metrics import accuracy_score
```

```python

def modelperformance(predictions):
    print('Accuracy score of model is {}'.format(accuracy_score(Y_test, predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
```

```python
 lor.fit(X_train,Y_train)
```

```python
lor_pred=lor.predict(X_test)
```

```python
modelperformance(lor_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
knn = KNeighborsClassifier()
paramknn={'n_neighbors':[3,5,7,],
         "weights":["uniform","distance"]}
gridknn = GridSearchCV(knn,paramknn)
gridknn.fit(X_train,Y_train)
```

```python
gridknn.best_params_
```

```python
knnpred = gridknn.predict(X_test)
```

```python
modelperformance(knnpred)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(probability=True)  
pramsvc = {"C":[0.01,0.1,0.02,0.03,0.04,0.1],
          "kernel":['linear','poly','rbf']}
```

```python
gridsvm = GridSearchCV(svm,pramsvc)
```

```python
gridsvm.fit(X_train,Y_train)
```

```python
pred_svc = gridsvm.predict(X_test)
modelperformance(pred_svc)
```

```python
joblib.dump(gridsvm,'loan_approval_model.pkl')
```
